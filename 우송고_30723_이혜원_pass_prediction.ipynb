# **데이터 전처리**

값이 너무 커서 0.01을 곱함

학습률 조정

레이어 3개 쌓음

```python
np.random.seed(20220214)
np.random.shuffle(data)
train_data = data[:380, :]
test_data = data[380:, :]

inputs = train_data[:, 0:23]*0.01
targets = train_data[:, -1:]

test_inputs = test_data[:,0:23]*0.01
test_targets = test_data[:, -1:]
```

```python
W1 = np.random.randn(16, 23)
B1 = np.random.randn(16, 1)
W2 = np.random.randn(8, 16)
B2 = np.random.randn(8, 1)
W3 = np.random.randn(1, 8)
B3 = np.random.randn(1, 1)

def forward(input, target):

    X = np.transpose(input, (1, 0))

    G1 = np.dot(W1, X) + B1

    R1 = np.maximum(0, G1)

    G2 = np.dot(W2, R1) + B2

    R2 = np.maximum(0, G2)

    G3 = np.dot(W3, R2) + B3

    sigmoid = 1 / (1 + np.exp(-G3))

    pred = np.transpose(sigmoid, (1, 0))

    loss = np.sum(-target*np.log(pred) - (1-target)*np.log(1-pred))

    return G1, R1, G2, R2, G3, sigmoid, pred, loss

_, _, _, _, _, _, pred, loss = forward(inputs, targets)
print('before loss', loss)

```

```python
def loss_gradient(input, target):
    G1, R1, G2, R2, G3, sigmoid, _, _ = forward(input, target)

    target = np.transpose(target, (1, 0))

    dL_dS = -1*( (target / sigmoid ) - ( (1-target) / (1-sigmoid)))

    dS_dG3 = ( 1/(1+np.exp(-G3))) * (1 - 1/(1+np.exp(-G3)))

    dG3_dR2 = np.transpose(W3, (1, 0))

    dG3_dW3 = np.transpose(R2, (1, 0))

    dG3_dB3 = np.ones_like(G3)

    dR2_dG2 = np.where(R2>0, 1, 0)

    dG2_dR1 = np.transpose(W2, (1, 0))

    dG2_dW2 = np.transpose(R1, (1, 0))

    dG2_dB2 = np.ones_like(G2)

    dR1_dG1 = np.where(R1>0, 1, 0)

    dG1_dW1 = input

    dG1_dB1 = np.ones_like(G1)



    dL_dG3 = dL_dS * dS_dG3

    dL_dW3 = np.dot(dL_dG3, dG3_dW3)

    dL_dB3 = np.sum(dL_dG3, keepdims=True) * dG3_dB3


    dL_dG2 = np.dot(dG3_dR2, dL_dG3) * dR2_dG2

    dL_dW2 = np.dot(dL_dG2, dG2_dW2)

    dL_dB2 = np.sum(dL_dG2, axis = 1, keepdims=True) * dG2_dB2


    dL_dG1 = np.dot(dG2_dR1, dL_dG2) * dR1_dG1

    dL_dW1 = np.dot(dL_dG1, dG1_dW1)

    dL_dB1 = np.sum(dL_dG1, axis=1, keepdims=True) * dG1_dB1

    return dL_dW1, dL_dB1, dL_dW2, dL_dB2, dL_dW3, dL_dB3

_, _, _, _, _, _, pred, loss = forward(inputs, targets)
print('before loss', loss)

learning_rate = 0.06
epochs = 5000

arr_loss = []
for i in range(epochs):
    SGD_num = np.random.randint(train_data.shape[0])
    x = inputs[SGD_num].reshape(1, -1)
    y = targets[SGD_num].reshape(1, -1)
    _, _, _, _, _, _, _, loss = forward(x, y)
    dL_dW1, dL_dB1, dL_dW2, dL_dB2, dL_dW3, dL_dB3 = loss_gradient(x, y)
    W1 = W1 + -1*learning_rate * dL_dW1
    B1 = B1 + -1*learning_rate * dL_dB1
    W2 = W2 + -1*learning_rate * dL_dW2
    B2 = B2 + -1*learning_rate * dL_dB2
    W3 = W3 + -1*learning_rate * dL_dW3
    B3 = B3 + -1*learning_rate * dL_dB3
    arr_loss.append(loss)


_, _, _, _, _, _, pred, loss = forward(inputs, targets)
print('after loss', loss)
```

```python
_, _, _, _, _, _, test_pred, test_loss = forward(test_inputs, test_targets)
print('test loss', test_loss)
print('test_pred', test_pred)
print('test_target', test_targets)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

torch.manual_seed(220214)

np.random.shuffle(data)

train_data = torch.FloatTensor(data[:380, :])
test_data = torch.FloatTensor(data[380:, :])

inputs = train_data[:, 0:23]
targets = train_data[:, -1:]

test_inputs = test_data[:, 0:23]
test_targets = test_data[:, -1:]

# 아래 Layer의 노드 수는 임의로 조절 가능함. 여러분들이 잘 최적화 시켜보세요.
model = nn.Sequential(
    nn.Linear(23, 32),
    nn.ReLU(),
    nn.Linear(32, 16),
    nn.ReLU(),
    nn.Linear(16, 8),
    nn.ReLU(),
    nn.Linear(8, 1),
    nn.Sigmoid()

)

pred = model(inputs)
loss = F.mse_loss(pred, targets)
print('before loss', loss)

#학습률(lr)과 학습 횟수(epochs)도 잘 조절하세요.
optimizer = optim.SGD(model.parameters(), lr=0.05)
epochs = 1000

arr_loss = []
for epoche in range(epochs + 1):
    pred = model(inputs)

    loss = F.mse_loss(pred, targets)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    arr_loss.append(loss.detach().numpy())

pred = model(inputs)
loss = F.mse_loss(pred, targets)
print('after loss', loss)

test_pred = model(test_inputs)
plt.plot(test_targets, 'ro', label='target')
plt.plot(test_pred.detach().numpy(), 'bo', label='pred')
plt.legend()
plt.show()

plt.plot(arr_loss)
plt.xlabel('epochs')
plt.ylabel('losses')
plt.show()
```
